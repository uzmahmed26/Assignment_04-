```python
from typing import List, Union

print("Lists and Dicts")

def access_element(my_list: List[str], index: int) -> str:
    try:
        return f"The element at index {index} is: {my_list[index]}"
    except IndexError:
        return "Invalid index! Please try again."

def modify_element(my_list: List[str], index: int, new_value: str) -> str:
    try:
        old_value = my_list[index]
        my_list[index] = new_value
        return f"Element at index {index} changed from '{old_value}' to '{new_value}'."
    except IndexError:
        return "Invalid index! Please try again."

def slice_list(my_list: List[str], start: int, end: int) -> str:
    try:
        # Validate slice range
        if 0 <= start < len(my_list) and 0 <= end <= len(my_list) and start < end:
            sliced = my_list[start:end]
            return f"Sliced list: {sliced}"
        else:
            return "Invalid slice indices! Please enter valid start and end indices."
    except Exception as e:
        return f"An error occurred: {e}"

def list_game() -> None:
    print("\nWelcome to the list manipulation game!")

    my_list = ["Apple", "Mango", "Orange", "Pear", "Peach"]

    while True:
        print("\nCurrent list:", my_list)
        print("Select an operation:")
        print("1. Access Element")
        print("2. Modify Element")
        print("3. Slice List")
        print("4. Quit")

        choice = input("Enter your choice (1-4): ")

        if choice == "1":
            try:
                index = int(input("Enter the index of the element you want to access: "))
                print(access_element(my_list, index))
            except ValueError:
                print("Please enter a valid integer for the index.")

        elif choice == "2":
            try:
                index = int(input("Enter the index of the element you want to modify: "))
                new_value = input("Enter the new value for the element: ")
                print(modify_element(my_list, index, new_value))
            except ValueError:
                print("Please enter a valid integer for the index.")

        elif choice == "3":
            try:
                start = int(input("Enter the start index for the slice: "))
                end = int(input("Enter the end index for the slice: "))
                print(slice_list(my_list, start, end))
            except ValueError:
                print("Please enter valid integers for the slice indices.")

        elif choice == "4":
            print("Exiting the game, thanks for playing!")
            break

        else:
            print("Invalid choice! Please enter a number between 1 to 4.")

# Run the game
list_game()
```
